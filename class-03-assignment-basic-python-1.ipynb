### 4. Calculate the Sum of Digits
**Task**: Given a number, find the sum of its digits.  
**Demo Input**: `123`  
**Demo Output**: `6`  
(Explanation: `1+2+3=6`)

```python
a  = 123
total = 0
for i in str(a):
    total = total+int(i)
print(total)

    
```

### 5. Find Factorial of a Number
**Task**: Calculate the factorial of a given number.  
**Demo Input**: `4`  
**Demo Output**: `24`  
(Explanation: `4! = 4*3*2*1 = 24`)

```python
a = 4
fact = 1
for i in range(a):
    i = i+1
    fact = fact*i
print(fact)
```

### 7. Find the GCD of Two Numbers
**Task**: Find the greatest common divisor (GCD) of two numbers.  
**Demo Input**: 8 and 12 <br>
**Demo Output**: `4`

```python
a = 8
b = 12
b_value = 0
s_value = 0
GCD = 0
if a>b:
    b_value = b_value+a
    s_value = s_value+b
else:
    b_value = b_value+b
    s_value = s_value+a
while b_value%s_value != 0:
    f = b_value%s_value
    b_value = s_value
    s_value = f
    GCD = s_value
print(GCD)
    
```

### 9. Check Armstrong Number
**Task**: Check if a number is an Armstrong number (the sum of its digits raised to the power of the number of digits equals the number).  
**Demo Input**: `153`  
**Demo Output**: `153 is an Armstrong Number`  
(Explanation: \(1^3 + 5^3 + 3^3 = 153\))

```python
a = 153
length = len(str(a))
total = 0
for i in str(a):
    total = total+ int(i)**length
if total == a:
    print(a,"is an Armstrong Number")
else:
    print(a,"is not an Armstrong Number")
```

### 10. Generate a Pattern
**Task**: Print a pyramid pattern with `n` rows.  
**Demo input:** 4 <br>
**Output:** You will see the pyramid pattern with 4 rows

```python
n = 4
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("*" * (2 * i - 1))
```
